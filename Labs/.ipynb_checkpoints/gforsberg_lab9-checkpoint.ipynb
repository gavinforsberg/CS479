```python
class BankAccount(object): 
    balance = 0
    ## Start your self.balance off at 0 dollars
    ## If the the pin is invalid for any method, tell the user invalid pin (merely print it)
    def __init__(self, pin):
        self.pin_str = pin
        self.balance = 0
        
        # Initial account self.balance is 0 and pin is your starting pin. If pin is not a 4 digit number ask
        # the user to enter a new pin
        # Enter pin as a string (otherwise 0123 will truncate to 123)
        digits = '1234567890'
        while len(self.pin_str) != 4:
            self.pin_str = input('Enter a new FOUR digit pin: ')

        ## Make sure only digits are allowed!!
        for i in range(0,4):
            if self.pin_str[i] not in digits:
                self.pin_str = input('Enter a new 4 digit pin (only digits): ')

        ## Your pin may not be the same digit repeated (1111 = bad, 1112 = ok)
        for i in range(1,4):
            if self.pin_str[i] == self.pin_str[0]:
                self.pin_str = input('Same digit repeated: Enter a new 4 digit pin: ')

    #Increment account self.balance by amount and return updated self.balance
    def deposit(self, pin, amount): 
        if pin == self.pin_str:
            self.balance += amount
            return(self.balance)
        else: 
            print('Invalid Pin\n')
            return(None)
            
    #Decrement account self.balance by the amount and return the amount withdrawn
    
    #If the user tries to withdraw more money than they have, print out they cannot withdraw that much money
    ##Do not withdraw any money at all (don't try to withdraw the difference) and return 0 for amount withdrawn
    def withdraw(self, pin, amount): 
        if pin == self.pin_str:
            if amount > self.balance:
                print('Cannot withdraw that much money.\n')
                return(0)
            else:
                self.balance -= amount
                return(amount)
        else: 
            print('Invalid Pin\n')
            return(None)
            

    #Return account self.balance
    def get_balance(self, pin): 
        if pin == self.pin_str:
            return self.balance
        else: 
            print('Invalid Pin\n')
            return(None)
    
    ##Updates the pin on the bank account, make sure new pin conforms to save rules from the constructor!
    ##Also your new pin may not be equal to your old pin
    def update_pin(self, oldpin, newpin): 
        if newpin:
            if oldpin == newpin:
                print('Invalid Pin: New pin cannot be equal to old pin.')
                return(None)
            else:
                self.pin_str = newpin
                return(None)
```

Now you will build a high interest yielding bank account!

Simply inherit from your BankAccount class with a few changes

Call this class HighYieldBankAccount

1)  Your constructor will need a percent attached to it as an input (self,pin,interest_rate) but needs to work identically to the constructor from the BankAccount class otherwise

2)  Calculate the balance of your account in t years from now using this handy dandy formula from high school!  Return that balance but **DO NOT** update the balance (this is a projection essentially)

**This method will be called compound_interest with parameters self,pin,years,compound_rate**

$P = C(1 + \frac{r}{n})^{nt}$

Where C = your balance when you start

n = number of times this is compounded per year (the compound rate)

t = number of years

r = interest rate from the constructor of the class (leave it as an integer, 6 means 6%)

If we let n tend towards infinity we wind up with the well known $P=Ce^{rt}$ equation for compound interest!  If you choose a very large number for n, it should converge towards that result!

```python
## Test your class
## Any assertion errors means something is wrong, you should see "invalid pin" twice below the cell

ba = BankAccount("1337")
assert(ba.get_balance("1337")==0)
assert(ba.withdraw("1337",400)==0) #No money can't withdraw
assert(ba.deposit("1337",1000)==1000) ## 0 -> 1000
assert(ba.deposit("1337",1000)==2000) ## 1000 -> 2000
assert(ba.withdraw("1337",1200)==1200) ## took out 1200, balance now 800
assert(ba.withdraw("1337",1200)==0) ## try to take out 1200, fail since balance is 800
assert(ba.get_balance("1337")==800) ## Check our balance is 800
assert(ba.deposit("1336",400)==None) ## Try to add 400, but we screwed the pin up (remember default return None)
assert(ba.update_pin("1337","1337")==None) #Cannot change pin to same pin
assert(ba.update_pin("1337","4242")==None) #This is fine
assert(ba.deposit("4242",1200)==2000) #Deposit 1200 dollars into the account (note pin is changed)
```

```python
## Your HighYieldBankAccount code goes here
## You will most likely need to rewrite the init function and update pin function (you'll see why)
## Other functions you should seamlessly inherit from the BankAccount class above

class HighYieldBankAccount(BankAccount):
    def __init__(self, pin, interest_rate):
        self.pin_str = pin
        self.balance = 0
        self.interest_rate_num = interest_rate
        
        # Initial account self.balance is 0 and pin is your starting pin. If pin is not a 4 digit number ask
        # the user to enter a new pin
        # Enter pin as a string (otherwise 0123 will truncate to 123)
        digits = '1234567890'
        while len(self.pin_str) != 4:
            self.pin_str = input('Enter a new FOUR digit pin: ')

        ## Make sure only digits are allowed!!
        for i in range(0,4):
            if self.pin_str[i] not in digits:
                self.pin_str = input('Enter a new 4 digit pin (only digits): ')

        ## Your pin may not be the same digit repeated (1111 = bad, 1112 = ok)
        for i in range(1,4):
            if self.pin_str[i] == self.pin_str[0]:
                self.pin_str = input('Same digit repeated: Enter a new 4 digit pin: ')
      
    def compound_interest(self, pin, years, compound_rate):
        return( self.balance * (1 + ((self.interest_rate_num / 100) / compound_rate)) ** (compound_rate * years) )
    
    def update_pin(self, oldpin, newpin): 
        if newpin:
            if oldpin == newpin:
                print('Invalid Pin: New pin cannot be equal to old pin.')
                return(None)
            else:
                self.pin_str = newpin
                return(None)
```
